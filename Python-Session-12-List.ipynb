**append** vs **extend** vs **concat**

```python
- append will add the element at last, last element can be any type

    - list1=[1,2,3]

    - list2=['A','B']

    - list1.append(list2)=[1,2,3,['A','B']]

- Extend will update the list by adding new elements, the result also save in a same list

    - list1=[1,2,3]

    - list2=['A','B']

    - list1.extend(list2)=[1,2,3,'A','B']

- Concat will add two list, but the result will update in a same? list

    - list1=[1,2,3]

    - list2=['A','B']

    - list1+list2=[1,2,3,'A','B']

                    
```

```python
#WAP take 5 random numbers
# and create a dictionary with even and odd

# Output: {'even':[20,22],'odd':[19,21,23]}

# step-1: take empty dictionary
# step-2: take two even and odd list
# step-3: import random
# step-4: for i in range(5):
# step-5:     num=random.randint(a,b)
# step-6:     if <even>:
# step-7:        append the values in even list
# step-8:    else:
# step-9:        append the values in odd list
# step-10: create dictinary
```

```python
import random

dic = {}
even = []
odd = []

for i in range(5):
    num = random.randint(1,100)
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

dic['even'] = even
dic['odd'] = odd

print(dic)
```

```python
import random

nums = [random.randint(1,100) for _ in range(5)]

dic = {
    'even': [n for n in nums if n % 2 == 0],
    'odd': [n for n in nums if n % 2 != 0]
}

print(nums)
print(dic)
```

```python
s1 = 'virat.kholi@blr.com,Rohit.sharma@mumbai.com,kl.rahul@lucknow.com'
emails = s1.split(',')

def inf(strin):
    name, company = strin.split('@')
    first, second = name.split('.')
    return first, second, company

dic = {
    "first_name": [],
    "second_name": [],
    "company": []
}

for e in emails:
    f, s, c = inf(e)
    dic["first_name"].append(f)
    dic["second_name"].append(s)
    dic["company"].append(c)

print(dic)
```

**if-else**

```python
# Create a function for finding greater number between two numbers
def greater(n1,n2):
    if n1>n2:
        return(n1)
    else:
        return(n2)
        
greater(100,200)
```

```python
l1=[]
def greater(n1,n2):

    if n1>n2:   
        l1.append(n1)
    else:
        l1.append(n2)
         
greater(100,200)
```

```python
l1=[<if_output> <if_condition>]
```

```python
l1=[]
def greater(n1,n2):

    if n1>n2:   
        l1.append(n1)
    else:
        l1.append(n2)
         
greater(100,200)

# format :
# lambda <arg1>,<arg2>: <if_output> <if_con> else <else_op>
```

```python
greater= lambda a,b: (a if a>b else b)
greater(8,3) 
```

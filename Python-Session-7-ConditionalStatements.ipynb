```python
Gender=input("Enter your gender:")
Age=eval(input("Enter your age:"))
if(Gender=='M'):
    if(Age>=60):                 
        print("Senior Citizen")          
    elif(Age>=30):
        print("Middle Aged Man")
    elif(Age>=20):
        print("Young boy ")
    elif(Age>=13):
        print("Teenage boy")
    else: 
        print("Boy")
elif Gender=="F":
    if(Age>=60):                 
        print("Senior Citizen")          
    elif(Age>=30):
        print("Middle Aged Woman")
    elif(Age>=20):
        print("Young Girl ")
    elif(Age>=13):
        print("Teenage Girl")
    else: 
        print("Girl")
else:
    print("Provide proper gender")
```

```python
def greet():
    print("hello")

def name():
    print('python')
    greet()
    
name()

# step-1: define greet()
# step-2: define name()
# step-3: calling name()
# step-4: python
# step-5: calling greet()
# step-6: hello
```

```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)

def main():
    print("if you want enter 1: addition operation willcome")
    print("if you want enter 2: subtraction operation willcome")
    print("if you want enter 3: multiplication operation willcome")
    print("if you want enter 4: division operation willcome")
    operation=input("Enter a operation number")
    a=eval(input('first number'))
    b=eval(input('second number'))
    if operation=='1':
        add(a,b)
    if operation=='2':
         sub(a,b)
    if operation=='3':
         mul(a,b)
    if operation=='4':
         div(a,b)

main()
```
